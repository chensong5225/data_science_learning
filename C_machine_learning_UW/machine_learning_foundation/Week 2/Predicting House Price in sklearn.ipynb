```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sales = pd.read_csv('home_data.csv')
sales.head()
```

```python
sales.describe()
```

```python
from sklearn.model_selection import train_test_split
target = sales.price
x_train, x_test, y_train, y_test = train_test_split(sales, target, train_size=0.8, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import math
x_train_sqft = x_train.sqft_living
x_test_sqft = x_test.sqft_living
sqft_model = LinearRegression().fit(x_train_sqft.reshape(-1, 1), y_train)
y_pred = sqft_model.predict(x_test_sqft.reshape(-1, 1))
print math.sqrt(mean_squared_error(y_test, y_pred))
```

```python
plt.plot(x_test_sqft,y_test,'.', x_test_sqft,y_pred,'-')
```

```python
print sqft_model.intercept_, sqft_model.coef_
```

```python
my_features = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'zipcode']
x_train_my = x_train[my_features]
x_test_my = x_test[my_features]

my_model = LinearRegression().fit(x_train_my, y_train)
y_pred = my_model.predict(x_test_my)
print math.sqrt(mean_squared_error(y_test, y_pred))
```
